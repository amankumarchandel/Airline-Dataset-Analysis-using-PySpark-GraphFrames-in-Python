## GraphFrames
- Graph Analysis
 - method used in computational biology
 - prominent data analysis technique for both social network analysis and recommender systems
 - Facebook friend cluster maps is a simple and intuitive example
- Getting started
 - GraphFrame is always created from a vertex and an edges DataFrame
 - Rules creating a GraphFrame
   - Vertex DataFrame must contain a column named id
   - Edges DataFrame must contain a column named src
- Directed vs undirected edges
 - Directed graphs have edges with no direction
 - Undirected graphs have edges with no direction
- Filtering and connected components
- Motif finding
 - Useful to discover structural patterns in graphs
 - find-function is the key to invoke the query
 - <a href="https://graphframes.github.io/graphframes/docs/_site/user-guide.html#motif-finding">Documentation</a>
- TriangleCount and PageRank Algorithms
 - TriangleCount
   - number of triangles passing through each vertex
   - triangle can be defined as a group of three vertices that is interrelated
 - PageRank
   - Initially used to know how important a website is
